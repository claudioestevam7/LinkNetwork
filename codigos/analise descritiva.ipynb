```python
import pandas as pd
import numpy as np

# Função para calcular estatísticas descritivas apenas para colunas numéricas
def calcular_estatisticas(dados):
    dados_numericos = dados.select_dtypes(include=[np.number])  # Seleciona apenas colunas numéricas
    
    estatisticas = {
        'Média': dados_numericos.mean(),
        'Mediana': dados_numericos.median(),
        'DP': dados_numericos.std(),
        'CV': dados_numericos.std() / dados_numericos.mean(),
        'Mínimo': dados_numericos.min(),
        'Máximo': dados_numericos.max(),
        'Amplitude': dados_numericos.max() - dados_numericos.min()
    }
    
    estatisticas_df = pd.DataFrame(estatisticas)
    return estatisticas_df

# Exemplo de uso
dados = pd.read_excel('data/2020.xlsx')  # Substitua pelo caminho do seu arquivo
estatisticas_df = calcular_estatisticas(dados)

# Para visualizar a tabela
print(estatisticas_df)

# Para salvar a tabela em um arquivo Excel, se necessário
estatisticas_df.to_excel('estatisticas_descritivas.xlsx')
```

```python
import pandas as pd
import numpy as np

# Função para calcular estatísticas descritivas apenas para colunas numéricas
def calcular_estatisticas(dados):
    dados_numericos = dados.select_dtypes(include=[np.number])  # Seleciona apenas colunas numéricas
    
    estatisticas = {
        'Média': dados_numericos.mean(),
        'Mediana': dados_numericos.median(),
        'DP': dados_numericos.std(),
        'CV': dados_numericos.std() / dados_numericos.mean(),
        'Mínimo': dados_numericos.min(),
        'Máximo': dados_numericos.max(),
        'Amplitude': dados_numericos.max() - dados_numericos.min()
    }
    
    estatisticas_df = pd.DataFrame(estatisticas)
    return estatisticas_df

# Exemplo de uso
dados = pd.read_excel('data/2020.xlsx')  # Substitua pelo caminho do seu arquivo
estatisticas_df = calcular_estatisticas(dados)

# Para visualizar a tabela
print(estatisticas_df)

# Função para converter DataFrame para LaTeX
def dataframe_to_latex(df, filename='tabela_estatisticas.tex'):
    with open(filename, 'w') as f:
        f.write(df.to_latex(index=True, header=True, float_format="%.2f"))

# Salvar a tabela em um arquivo LaTeX
dataframe_to_latex(estatisticas_df)
```
