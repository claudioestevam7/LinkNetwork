```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from mplfinance.original_flavor import candlestick_ohlc

# Coletando os dados da Vivo (VIVT3.SA) para o ano de 2018
ticker = "VIVT3.SA"
start_date = "2018-01-01"
end_date = "2018-12-31"
data = yf.download(ticker, start=start_date, end=end_date)

# Preparando os dados para o grfico de velas
data['Date'] = data.index
data['Date'] = data['Date'].map(mdates.date2num)
ohlc = data[['Date', 'Open', 'High', 'Low', 'Close']]

# Plotando o grfico de velas
fig, ax = plt.subplots(figsize=(20, 6))

ax.xaxis_date()
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
ax.xaxis.set_major_locator(mdates.MonthLocator())

candlestick_ohlc(ax, ohlc.values, width=0.6, colorup='g', colordown='r', alpha=0.8)

plt.title('VIVT3 (2018)', fontsize=20)
plt.xlabel('Data', fontsize=15)
plt.ylabel('Preo (R$)', fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True)

plt.show()
```

```python
import yfinance as yf
import pandas as pd
import plotly.graph_objects as go

# Coletando os dados da Embraer (EMBR3.SA) para o ano de 2019
ticker = "EMBR3.SA"
start_date = "2019-01-01"
end_date = "2019-12-31"
data = yf.download(ticker, start=start_date, end=end_date)

# Plotando o grfico de velas com Plotly
fig = go.Figure(data=[go.Candlestick(
    x=data.index,
    open=data['Open'],
    high=data['High'],
    low=data['Low'],
    close=data['Close'],
    increasing_line_color='green', 
    decreasing_line_color='red'
)])

fig.update_layout(
    title='Grfico de Velas - EMBR3 (2019)',
    yaxis_title='Preo (R$)',
    xaxis_title='Data',
    xaxis_rangeslider_visible=False,
    template='plotly_dark'
)

fig.show()
```

```python
import yfinance as yf
import pandas as pd
import plotly.graph_objects as go
import plotly.offline as pyo

# Configurando o modo offline do Plotly
pyo.init_notebook_mode(connected=True)

# Coletando os dados da Embraer (EMBR3.SA) para o ano de 2019
ticker = "EMBR3.SA"
start_date = "2019-01-01"
end_date = "2019-12-31"
data = yf.download(ticker, start=start_date, end=end_date)

# Plotando o grfico de velas com Plotly
fig = go.Figure(data=[go.Candlestick(
    x=data.index,
    open=data['Open'],
    high=data['High'],
    low=data['Low'],
    close=data['Close'],
    increasing_line_color='green', 
    decreasing_line_color='red'
)])

fig.update_layout(
    title='Grfico de Velas - EMBR3 (2019)',
    yaxis_title='Preo (R$)',
    xaxis_title='Data',
    xaxis_rangeslider_visible=False,
    template='plotly_dark'
)

# Renderizando o grfico em um arquivo HTML e abrindo no navegador
pyo.plot(fig, filename='embraer_candlestick.html')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some sample data
np.random.seed(0)
cluster_1 = np.random.normal(0, 1, (300, 2)) + np.array([0, 0])
cluster_2 = np.random.normal(0, 1, (300, 2)) + np.array([5, -2])
cluster_3 = np.random.normal(0, 1, (300, 2)) + np.array([5, 7])

data = np.vstack([cluster_1, cluster_2, cluster_3])

# Define colors and cluster centers
colors = ['blue', 'orange', 'green']
centers = [np.mean(cluster_1, axis=0), np.mean(cluster_2, axis=0), np.mean(cluster_3, axis=0)]

# Plot the clusters
plt.figure(figsize=(8, 6))
# plt.title('Ideal Clustering')
plt.scatter(cluster_1[:, 0], cluster_1[:, 1], c='#080452', s=20)
plt.scatter(cluster_2[:, 0], cluster_2[:, 1], c='#3089b0', s=20)
plt.scatter(cluster_3[:, 0], cluster_3[:, 1], c='#943b07', s=20)

# Plot the centers with dark blue color
for center in centers:
    plt.scatter(center[0], center[1], c='darkblue', s=100)

# plt.xlim(-4, 10)
# plt.ylim(-4, 10)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Configuraes
n_ts_per_blob = [1,3] # Nmero de sries temporais por grupo
sz = 50  # Comprimento das sries temporais

# Gerar sries temporais simuladas
np.random.seed(0)
time = np.arange(sz)
series = [np.sin(0.1 * time + np.random.uniform(0, 2 * np.pi)) + 0.1 * np.random.randn(sz) for _ in n_ts_per_blob]

# Plotar as sries temporais geradas
plt.figure(figsize=(10, 6))
for i, serie in enumerate(series):
    plt.plot(serie, label=f'Srie Temporal {i+1}')
plt.title("Sries Temporais Similares")
plt.xlabel("Tempo")
plt.ylabel("Valor")
plt.legend()
plt.show()

```
